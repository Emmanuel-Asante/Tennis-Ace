```python
# Import modules
#import codecademylib3_seaborn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
from scipy.stats import pearsonr
```

```python
# load and investigate the data here:

# Load data into a dataframe
df = pd.read_csv("tennis_stats.csv")

# Inspect dataframe
df.head()
```

```python
# perform exploratory analysis here:

# Create a function
def exploratory_analysis_winnings():
    # Create list of game features
    game = ['Aces', 'DoubleFaults', 'FirstServe', 'FirstServePointsWon', 'SecondServePointsWon', 'BreakPointsFaced',
            'BreakPointsSaved', 'ServiceGamesPlayed', 'ServiceGamesWon', 'TotalServicePointsWon', 'FirstServeReturnPointsWon',
           'SecondServeReturnPointsWon', 'BreakPointsOpportunities', 'BreakPointsConverted', 'ReturnGamesPlayed', 'ReturnGamesWon',
           'ReturnPointsWon', 'TotalPointsWon']
    # Create an empty dictionary
    corr_dict = {}
    # Iterate through game feature list
    for item in game:
        # Create a scatter plot of a feature vs. 'Winnings'
        sns.scatterplot(x=df[item], y=df['Winnings'])
        # Create plot title
        plt.title(item + ' vs. Winnings')
        # Show plot
        plt.show()
        # Calculate correlation coefficient
        corr, p = pearsonr(df[item], df['Winnings'])
        # Update corr_dict with a game feature and its corresponding correlation coefficient
        corr_dict[item] = corr
        # Output corr
        print('The correlation between ' + item + ' and Winnings is ' + str(corr))
        # Clear current plot
        plt.clf()
    # Output corr_dict
    print(corr_dict)
    # Get the highest correlation coefficient with its corresponding game feature
    itemMaxValue = max(corr_dict.items(), key=lambda x : x[1])
    # Output correlation coefficient
    print('\nHeighest correlation coefficient between a feature and Winnings: ', itemMaxValue[1])
    # Output game feature
    print('Feature: ', itemMaxValue[0])
    
    
# Create a function    
def exploratory_analysis_losses():
    # Create list of game features
    game = ['Aces', 'DoubleFaults', 'FirstServe', 'FirstServePointsWon', 'SecondServePointsWon', 'BreakPointsFaced',
            'BreakPointsSaved', 'ServiceGamesPlayed', 'ServiceGamesWon', 'TotalServicePointsWon', 'FirstServeReturnPointsWon',
           'SecondServeReturnPointsWon', 'BreakPointsOpportunities', 'BreakPointsConverted', 'ReturnGamesPlayed', 'ReturnGamesWon',
           'ReturnPointsWon', 'TotalPointsWon']
    # Create an empty dictionary
    corr_dict = {}
    # Iterate through game feature list
    for item in game:
        # Create a scatter plot of a feature vs. 'Winnings'
        sns.scatterplot(x=df[item], y=df['Losses'])
        # Create plot title
        plt.title(item + ' vs. Losses')
        # Show plot
        plt.show()
        # Calculate correlation coefficient
        corr, p = pearsonr(df[item], df['Losses'])
        # Update corr_dict with a game feature and its corresponding correlation coefficient
        corr_dict[item] = corr
        # Output corr
        print('The correlation between ' + item + ' and Losses is ' + str(corr))
        # Clear current plot
        plt.clf()
    # Output corr_dict
    print(corr_dict)
    # Get the highest correlation coefficient with its corresponding game feature
    itemMaxValue = max(corr_dict.items(), key=lambda x : x[1])
    # Output correlation coefficient
    print('\nHeighest correlation coefficient between a feature and Losses: ', itemMaxValue[1])
    # Output game feature
    print('feature: ', itemMaxValue[0])   
```

```python
# Function call
exploratory_analysis_winnings()
```

```python
# Function call
exploratory_analysis_losses()
```

```python
# Analysis from the above scatter plot
print("There exist a strong positive relationship between ServiceGamesPlayed feature and Winnings outcome")
```

```python
## perform single feature linear regressions here:

# Select FirstServeReturnPointsWon feature
features = df[['FirstServeReturnPointsWon']]

# Select winning outcome
outcome = df[['Winnings']]

# Split data into training and test sets
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8, test_size = 0.2)

# Create a linear regression model
model = LinearRegression()

# Train model on training data
model.fit(features_train, outcome_train)

# Score model on test data
print('Predicting Winnings with BreakPointsOpportunities Test Score:', model.score(features_test, outcome_test))

# Find predicted outcome
prediction = model.predict(features_test)

# Create a scatter plot of predicted vs. outcome_test
plt.scatter(outcome_test, prediction, alpha=0.4)

# Show plot
plt.show()
```
